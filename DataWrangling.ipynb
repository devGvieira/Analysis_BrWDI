```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('Data.csv')

data.head()
```

```python
columns = data.columns
print(columns)
```

```python
columns_drop = ['Country Name', 'Country Code', 'Series Code']
data = data.drop(columns_drop, axis=1)
data.head()
```

```python
data = data.replace('..', np.nan)
data.info()
```

```python
data.head()
```

```python
data = data.dropna(subset=data.columns[1:] ,axis=0, how='all')
data.head()
```

```python
df = data.transpose()
df.head()
```

```python
range_years = ['Years', 1990, 2000]

for year in range(2013,2023):
    range_years.append(int(year))

years = range_years

df.insert(0, '-1', years)

df.head()
```

```python
df.reset_index(drop=True, inplace=True)
df.head()
```

```python
header = [num for num in range(len(df.columns))]

df.columns = header

df.head()
```

```python
df.to_csv('Data_Wrangled.csv')
```
